```python
# Função para preencher os CPFs com zeros a esquerda
# CPF tem sempre 11 digitos
```

```python
def zeros(linha):
    
    cpf = linha['CPF_Político_Autor*']
    
    try:
        int(cpf)
    except ValueError:
        return linha['CPF_Político_Autor*']
        
    numero = str(cpf.strip())
    numero = numero.replace(',','')
    numero = numero.replace('.','')
    numero = numero.replace('-','')
    numero = numero.replace('/','')
    #print(linha['Número_Processo'])

    if len(numero) < 11:
        conta = 11 - (len(numero))
        zero = '0'
        zeros = zero * conta
        alterado = zeros + numero
    elif len(numero) == 11:
        alterado = numero
    
    return alterado
```

```python
processos['CPF_Político_Autor*'] = processos.apply(zeros, axis=1)
```

```python
# Cruza a base do TSE com a do Ctrl+X
# Elimina CPF duplicados, caso existam no TSE
```

```python
processos_encontrados = pd.merge(candidatos_2020.drop_duplicates('cpf'), processos, left_on='cpf', right_on='CPF_Político_Autor*')
```
